## 1. Git 저장소 만들기


 1. 아직 버전관리를 하지 않는 로컬 디렉토리 하나를 선택해서 Git 저장소를 적용하는 방법
 2. 다른 어딘가에서 Git 저장소를 Clone 하는 방법
 
두 가지 중 한 가지 방법으로 Git 저장소를 쓸 수 있음

1. 기존 디렉토리를 Git 저장소로 만들기 <br>
 1) 프로젝트의 디렉터리로 이동함(운영체제마다 디폴트 위치가 다름) <br>
 Windows `$ cd /c/user/my_project` <br>
 Linux `$ cd /home/user/my_project` <br>
 Mac `$ cd /Users/user/my_project` <br>
 `cd`는 change directory를 뜻함 <br>
 
 2) init 명령어는 `.git`이라는 하위 디렉토리를 만들며 저장소에 필요한 중요 파일이 들어있음(만들어지는 파일은 추후 다루도록 함)<br>
 `$ git init`을 실행하면 git에 맞도록 디렉토리가 초기화됨<br>
 
 3) Git으로 파일을 관리하려면 저장소에 파일을 추가하고 커밋해야함 <br>
`$ git add *.c` 파일을 추가함<br>
`$ git add LICENSE`<br>
`$ git commit -m 'initial project version'` 파일을 커밋함


2. 기존 저장소를 Clone 하기
 * 다른 프로젝트에 참여하려거나 Git 저장소를 복사하고 싶을 때 `git clone`을 사용
 * Git이 Subversion등 시스템과 다른 점은 서버에 있는 거의 모든 데이터를 복사한다는 것
 * `git clone`을 실행하면 프로젝트 히스토리를 전부 받아옴 
<br><br>
 * `git clone <url>` 명령으로 저장소를 clone 함 <br>
 `$ git clone https://github.com/libgit2/libgit2` <br>
 "libgit2"라는 디렉토리를 만들고 그 안에 `.git` 디렉토리를 만듦 <br>
 저장소의 데이터를 모두 가져와서 자동으로 가장 최신 버전을 Checkout 해 놓음 <br>
 `libgit2` 디렉토리로 이동하면 Checkout으로 생성한 파일을 볼 수 있고 당장 작업 가능함

 * `$ git clone https://github.com/libgit2/libgit2 mylibgit` <br>
 디렉토리 이름이 `mylibgit` 으로 지정됨

<br>

## 2. 수정하고 저장소에 저장하기

파일을 수정하고 파일의 스냅샷을 커밋하기

* 파일의 세 가지 상태 (복습)
 * Committed: 데이터가 로컬 데이터베이스에 안전하게 저장됨
 * Modified: 수정한 파일을 아직 커밋하지 않은 상태
 * Staged: 현재 수정한 파일을 곧 커밋할 것이라고 표시한 상태
<br>
![파일의 세 가지 상태](fig1-6.png)
<center>워킹 트리, Staging Area, Git 디렉토리</center>
<br>

* 워킹 디렉토리의 파일은 Tracked(관리대상임) Untracked(관리대상이 아님) 으로 나눔
* Tracked 파일은 이미 스냅샷에 포함되어 있던 파일이며 Unmodified(수정하지 않음)와 Modified(수정함) 그리고 Staged(커밋으로 저장소에 기록할 상태) 으로 나눔
* Untracked 파일은 워킹 디렉토리에 있는 파일 중 스냅샷, Staging Area 어디에도 포함되지 않은 파일임
* 처음 저장소를 Clone 하면 모든 파일은 Tracked - Unmodified 상태(Checkout 이후 수정하지 않았으므로)
* 커밋 이후 파일을 처음으로 수정하면 Git은 그 파일을 Modified로 인식함
* 커밋을 하기 위해서는 이러한 Modified 파일을 Staged 상태로 만든 뒤에 Staged 상태인 파일을 커밋함

![파일의 라이프사이클](fig2-1.png)
<center>파일의 라이프사이클</center>
<br>

* 파일의 상태 확인하기
 * `git status` 명령으로 파일의 상태를 확인할 수 있음
 * Clone 후에 사용하면 Tracked 파일을 하나도 수정하지 않았다는 것을 말해줌 <br>
`On branch master
Your Branch is up-to-date with 'origin/master'.
nothing to commit, working directory clean`
 * 브랜치 이름이 master로 나오는데 브랜치는 추후 다루도록 함
 
 * 프로젝트에 `README`파일을 새로 만든 뒤 상태는 Untracked 일 것임<br>
`$ echo 'My Project' > README`<br>
`$ git status`<br>
`On branch master`<br>
`Your branch is up-to-date with 'origin/master'.`<br>
`Untracked files:`<br>
`  (use "git add <file>..." to include in what will be committed)` <br>
`    README` <br>
`nothing added to commit but untracked files present (use "git add" to track)`
 * `README` 파일은 "Untracked files"에 속해 있음
 * Git은 Untracked 파일을 아직 스냅샷(커밋)에 넣어지지 않은 파일이라고 봄


* 파일을 새로 추적하기
 * `git add` 명령으로 파일을 새로 추적할 수 있음(Tracked 상태로 바꿀 수 있음)
 * Tracked 상태가 되면서 커밋에 추가될 Staged 상태로 만들 수 있음
`$ git add README` <br>
`$ git status`<br>
`On branch master`<br>
`Your branch is up-to-date with 'origin/master'.`<br>
`Changes to be committed:`<br>
`  (use "git reset HEAD <file>..." to unstage)`<br>
`    new file:   README`
 * "Changes to be committed"는 커밋될 것 이므로 Staged 상태를 의미함
 * 커밋하면 `git add`를 실행한 시점의 파일이 커밋되어 저장소 히스토리에 남음
 * `git add`는 파일 또는 디렉토리의 경로를 인수로 받으므로 디렉토리를 추가하면 그 아래에 있는 모든 파일들까지 추가함
